1. What is Simple Linear Regression ?                                               
->  Simple Linear Regression is a statistical technique used to analyze the relationship between two variables. One variable is called the independent variable (X) and the other is called the dependent variable (Y). The main objective of simple linear regression is to predict the value of the dependent variable based on the independent variable.

      The relationship between the variables is represented by the following equation:

      Y = β₀ + β₁X

      Where:

      Y = Dependent variable (the value to be predicted)

      X = Independent variable (input or predictor variable)

      β₀ (Beta 0) = Intercept of the regression line. It represents the value of Y when X = 0.

      β₁ (Beta 1) = Slope or regression coefficient, which indicates how much Y changes when X increases by one unit.

      The regression model finds the best fitting straight line that minimizes the difference between the actual values and predicted values. These differences are called errors or residuals.

      Simple Linear Regression is widely used for prediction and analysis of relationships between variables, such as predicting sales based on advertising cost or predicting marks based on study hours.

2. What are the key assumptions of Simple Linear Regression ?                           
Simple Linear Regression works properly only when certain assumptions about the data are satisfied. The main assumptions are:

      1. Linearity
      There should be a linear relationship between the independent variable (X) and the dependent variable (Y). This means the change in Y is proportional to the change in X.

      2. Independence of Errors
       The residuals (errors) should be independent of each other. This means the error of one observation should not influence the error of another observation.

      3. Homoscedasticity (Constant Variance)
      The variance of the residuals should remain constant for all values of X. In other words, the spread of errors should be similar across all levels of the independent variable.

      4. Normal Distribution of Errors
      The residuals should be normally distributed. This assumption is important for performing statistical tests and confidence intervals.

      5. No Significant Outliers
      The dataset should not contain extreme values or outliers that strongly affect the regression line.

      If these assumptions are satisfied, the simple linear regression model provides accurate predictions and reliable statistical results.

7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?                                                                     

    The Coefficient of Determination (R²) in Simple Linear Regression measures how well the regression model explains the variation in the dependent variable (Y).

    It represents the proportion of the total variation in Y that is explained by the independent variable (X) through the regression model.

    The value of R² ranges from 0 to 1:

    - R² = 1 → The regression model perfectly explains all the variation in Y.

     - R² = 0 → The model does not explain any variation in Y.

    - 0 < R² < 1 → The model explains some portion of the variation in Y.

    For example, if R² = 0.75, it means that 75% of the variation in the dependent variable is explained by the independent variable, while the remaining 25% is due to other factors or random error.

    Thus, the coefficient of determination indicates the goodness of fit of the regression model and shows how well the model fits the observed data.

10. What are the key assumptions of Multiple Linear Regression ?                         
  
     Key Assumptions of Multiple Linear Regression

    Multiple Linear Regression works correctly only when certain assumptions are satisfied. The main assumptions are:

    - Linearity                                                                  
      There should be a linear relationship between the dependent variable (Y) and each independent variable (X₁, X₂, …).

    - Independence of Errors

      The residuals (errors) should be independent of each other. The error in one observation should not affect another observation.

     - Homoscedasticity (Constant Variance)

        The variance of the residuals should remain constant across all values of the independent variables.

      - Normal Distribution of Errors

        The residuals should be normally distributed, which is important for statistical inference and hypothesis testing.

      - No Multicollinearity

        The independent variables should not be highly correlated with each other, because strong correlation between predictors can distort the estimation of regression coefficients.

     - No Significant Outliers

       The data should not contain extreme outliers that can strongly influence the regression results.

      If these assumptions are satisfied, the multiple linear regression model can provide reliable predictions and accurate analysis of relationships between variables.

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model ?                                              

      Multiple Linear Regression model

      Heteroscedasticity refers to a situation in Multiple Linear Regression where the variance of the residuals (errors) is not constant across all levels of the independent variables.

       In a good regression model, the residuals should have constant variance, which is called homoscedasticity. However, when the spread of errors increases or decreases for different values of the predictors, it is called heteroscedasticity.

       Effect on the Regression Model:

      - It violates one of the key assumptions of multiple linear regression.

     - The estimated regression coefficients may still be unbiased, but they become less reliable.

     - It can lead to incorrect standard errors, which affects hypothesis testing and confidence intervals.

    - As a result, statistical tests like t-tests and F-tests may give misleading results.

     Thus, heteroscedasticity reduces the accuracy and reliability of the regression analysis, making it harder to draw correct conclusions from the model.

13. What are some common techniques for transforming categorical variables for use in regression models ?                                                         
In regression models, categorical variables (such as gender, color, or city) cannot be used directly because regression requires numerical input. Therefore, categorical variables must be transformed into numerical form. Some common techniques are:

     - Dummy Variable Encoding (One-Hot Encoding)

         Each category is converted into a separate binary variable (0 or 1). For example, a variable “Color” with categories Red, Blue, and Green can be converted into three dummy variables.

    - Label Encoding

         Each category is assigned a unique numerical value. For example:
         Male = 0, Female = 1.

    - Binary Encoding

      Categories are first converted into numbers and then represented in binary form, which reduces the number of columns compared to one-hot encoding.

     - Ordinal Encoding

       Used when categories have a natural order, such as Low, Medium, High. These are encoded with numbers like 1, 2, and 3.

      - Target Encoding
        Each category is replaced with the average value of the target variable for that category.

      These techniques help convert categorical data into numerical form so it can be used effectively in regression models.

18. What are the limitations of using R² as a sole measure of model performance ?                                                                   

    Limitations of Using R-square as the Only Measure of Model Performance

    R-square is a statistical measure used in regression analysis to show how much of the variation in the dependent variable is explained by the independent variables in the model. Although it is useful for understanding model fit, it has several limitations when used alone.

    First, a high R-square value does not always mean the model is good or accurate. It only indicates that the model explains a large portion of the variation in the data, but it does not confirm that the relationship between variables is correct or meaningful.

    Second, R-square increases when more independent variables are added to the model, even if those variables do not actually improve the prediction. This can make the model appear better than it really is.

    Third, R-square cannot detect problems in the regression model, such as multicollinearity, heteroscedasticity, or violations of regression assumptions.

    Fourth, R-square does not directly measure prediction performance on new or unseen data. A model may have a high R-square value but still give poor predictions.

    Therefore, R-square should not be used as the only measure of model performance. It should be evaluated along with other measures and diagnostic tests to properly judge the quality and reliability of a regression model.

20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?                                                         

    Identification and Importance of Heteroscedasticity in Residual Plots

    - Heteroscedasticity occurs when the variance of the residuals (errors) is not constant across all levels of the independent variables in a regression model. In a good regression model, the residuals should have a constant spread around zero.

    - Heteroscedasticity can be identified using a residual plot, where the residual values are plotted against the predicted values or independent variables. If the residuals form a clear pattern, such as a funnel shape, cone shape, or increasing and decreasing spread, it indicates heteroscedasticity. In contrast, if the residuals are randomly scattered with equal spread, it suggests that the variance is constant and the assumption of homoscedasticity is satisfied.

     - It is important to address heteroscedasticity because it violates one of the key assumptions of regression analysis. When this problem occurs, the estimated coefficients may still remain unbiased, but the standard errors become unreliable. As a result, statistical tests and confidence intervals may become inaccurate, which can lead to incorrect conclusions about the significance of variables.

     Therefore, detecting and correcting heteroscedasticity helps ensure more reliable statistical inference and more accurate interpretation of the regression model.

Interpretation of High R-square but Low Adjusted R-square in Multiple Linear Regression

    - If a Multiple Linear Regression model has a high R-square but a low adjusted R-square, it usually indicates that the model includes too many independent variables that do not significantly contribute to explaining the dependent variable.

     - R-square measures how much variation in the dependent variable is explained by the model. However, it always increases when more predictors are added, even if those predictors are not useful.

     - Adjusted R-square, on the other hand, penalizes the model for including unnecessary variables. It only increases when the added variables actually improve the model.

    Therefore, when R-square is high but adjusted R-square is low, it suggests that some independent variables are irrelevant or weak predictors. These variables increase the complexity of the model without improving its predictive ability.

    This situation may lead to overfitting, where the model fits the training data well but performs poorly on new data. To improve the model, unnecessary variables should be removed and only the most significant predictors should be retained.

    The general equation for polynomial regression can be written in one line as:

    Y = a + b1X + b2X² + b3X³ + … + bnXⁿ

    By increasing the degree of the polynomial, the regression model can capture more complex patterns and curved relationships in the data. This allows polynomial regression to provide a better fit when the relationship between variables is not linear.

28. What are the limitations of polynomial regression ?                                    

    Limitations of Polynomial Regression

    Polynomial regression is useful for modeling non-linear relationships, but it also has several limitations.

    One major limitation is overfitting. When the degree of the polynomial becomes very high, the model may fit the training data extremely well but perform poorly on new or unseen data. This reduces the model’s ability to generalize.

    Another limitation is that polynomial regression can become very complex and difficult to interpret. As more polynomial terms are added, it becomes harder to understand how each variable influences the dependent variable.

    Polynomial regression is also sensitive to extreme values or outliers. A few unusual data points can significantly change the shape of the curve, which may lead to inaccurate predictions.

    In addition, high-degree polynomial models may show large fluctuations at the edges of the data range, which can produce unrealistic predictions outside the observed data.

    Finally, polynomial regression can require more computational effort and careful model selection, because choosing the correct degree of the polynomial is important for achieving a good balance between accuracy and simplicity.

29. What methods can be used to evaluate model fit when selecting the degree of a polynomial ?                                                                     

    Methods to Evaluate Model Fit When Selecting the Degree of a Polynomial

    When choosing the degree of a polynomial in polynomial regression, it is important to evaluate how well the model fits the data while avoiding overfitting. Several methods can be used to assess the quality of the model.

    One common method is the coefficient of determination, which measures how well the model explains the variation in the dependent variable. A higher value generally indicates a better fit, but relying only on this measure can sometimes be misleading when the model becomes too complex.

    Another useful method is adjusted R square, which adjusts the goodness-of-fit measure by considering the number of predictors in the model. This helps prevent selecting a polynomial with too many terms that do not significantly improve the model.

    Residual analysis is also important. By examining the residual values (the differences between actual and predicted values), we can check whether the errors are randomly distributed. If patterns appear in the residuals, it may indicate that the chosen polynomial degree is not appropriate.

    Cross-validation is another technique used to evaluate model performance. The data is divided into training and testing sets, and the model is evaluated on unseen data to check whether it generalizes well.

    Finally, error metrics such as Mean Squared Error or Root Mean Squared Error can be used to compare models with different polynomial degrees. The model with lower prediction error is generally preferred.

    Using these methods together helps in selecting a polynomial degree that provides a good balance between accuracy and simplicity.
